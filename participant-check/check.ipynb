```python
import datetime as dt
import pandas as pd
dt.datetime.now()
```

```python
def participate(middle_list, beginner_list, participate_date, df):
    check_arr = []
    for p in list(df['닉네임']):
        if p in middle_list: check_arr.append('M')
        elif p in beginner_list: check_arr.append('B')
        else: check_arr.append('.')
    check_arr

    date = dt.datetime.strptime(participate_date, '%Y-%m-%d')
    df[date] = check_arr
```

```python
check_df = pd.read_csv('simple-check', index_col=0)
```

```python
participate(['흐느ㅎ', 'Ashe', '후플푸프'], [], '2022-10-10', check_df)
```

```python
check_df = check_df.set_index("닉네임")
```
